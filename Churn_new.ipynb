```python
## Loadig Dataset
from google.colab import drive
drive.mount('/content/drive',force_remount =True)
# Update the file path to include the full path from the drive's root directory.
df =pd.read_csv('/content/drive/MyDrive/data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
telecom = df.copy()
telecom["TotalCharges"] = pd.to_numeric(telecom["TotalCharges"], errors = 'coerce')
telecom.isnull().sum()
```

```python
telecom.loc[telecom["TotalCharges"].isnull() ==True]
```

```python
telecom.dropna(how ='any', inplace =True)
```

```python
telecom.TotalCharges.isnull()
telecom.isnull().sum()
```

```python
labels = [f"{i}- {i+11}" for i in range(1, 73,12)]
telecom['tenure_group'] = pd.cut(telecom['tenure'], range(1,74,12),right =False, labels = labels)
telecom.tenure_group.value_counts()
```

```python
telecom.drop(columns =["customerID", "tenure"],axis = 1, inplace= True)
```

```python
cat_cols = telecom.select_dtypes(include = 'object').columns
telecom[cat_cols].nunique()
for col in cat_cols:
  print(f"{col} value is {telecom[col].unique()}")
```

```python
# Label Encoding
binary_cols = ["Partner", "Dependents", "PhoneService", "PaperlessBilling", "Churn"]
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()
telecom_lab= telecom.copy()
for col in binary_cols:
  telecom_lab[col] = telecom[col].map({"Yes" :1, "No" : 0})

#bool_cols = telecom_ohe.select_dtypes(include='bool').columns

```

```python
telecom_lab["tenure_group"].dtype
```

```python
telecom_lab.drop(columns ="gender", inplace = True)
```

```python
telecom_lab.shape
```

```python


# Assume telecom is the original dataset
categorical_columns = telecom.select_dtypes(include=['object']).columns  # Select categorical columns

# Count unique values per categorical column
unique_counts = telecom[categorical_columns].nunique()

# Extract only multi-class categorical columns (ignore binary ones)
multi_class_cols = unique_counts[unique_counts > 2].index.tolist()

# Apply One-Hot Encoding for Multi-Class Columns
telecom_ohe = pd.get_dummies(telecom_lab, columns=multi_class_cols+['tenure_group'], drop_first=True)

# Verify Encoding

telecom_ohe.head()

bool_cols = telecom_ohe.select_dtypes(include='bool').columns
# Convert only boolean columns to int
telecom_ohe[bool_cols] = telecom_ohe[bool_cols].astype(int)

# Verify the data types
print(telecom_ohe.dtypes)
```

```python
telecom_ohe
```

```python
# Featue Scaling
#Identify Numerical Features
from sklearn.preprocessing import MinMaxScaler
features_scale =["MonthlyCharges", "TotalCharges"]

scaler = MinMaxScaler()
telecom_ohe[features_scale] = scaler.fit_transform(telecom_ohe[features_scale])
telecom_ohe
```

```python
# ============================
# 📦 SETUP: Tracking Config (KEEP)
# ============================
import mlflow
mlflow.set_tracking_uri("file:///content/mlruns")  # Local tracking directory in Colab
mlflow.set_experiment("Churn Models Comparison")

# ============================
# ✅ IMPORTS (KEEP)
# ============================
import mlflow.sklearn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, roc_curve
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
from tensorflow.keras import layers
from google.colab import files
from fpdf import FPDF

# ============================
# ✅ DEFINE CLASSICAL ML MODELS
# ============================
def get_classical_models():
    return {
        "Logistic_Regression": LogisticRegression(max_iter=1000),
        "Random_Forest": RandomForestClassifier(n_estimators=100, random_state=42),
        "Gradient_Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
        "XGBoost": XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss'),
        "LightGBM": LGBMClassifier(n_estimators=100),
        "SVM": SVC(probability=True),
        "Naive_Bayes": GaussianNB()
    }

# ============================
# ✅ DEFINE MULTIPLE KERAS MODELS
# ============================
def get_keras_model_v1(input_shape):
    model = keras.Sequential([
        layers.Input(shape=(input_shape,)),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(32, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def get_keras_model_v2(input_shape):
    model = keras.Sequential([
        layers.Input(shape=(input_shape,)),
        layers.Dense(128, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ============================
# ✅ DEFINE MULTIPLE PYTORCH MODELS
# ============================
class PyTorchModelV1(nn.Module):
    def __init__(self, input_dim):
        super(PyTorchModelV1, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 32)
        self.out = nn.Linear(32, 1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.out(x))
        return x

class PyTorchModelV2(nn.Module):
    def __init__(self, input_dim):
        super(PyTorchModelV2, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.dropout = nn.Dropout(0.2)
        self.fc2 = nn.Linear(128, 64)
        self.out = nn.Linear(64, 1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.out(x))
        return x

# ============================
# ✅ MODIFIED: Plot and Log with Hyperparams
# ============================
def plot_and_log_figures(model_name, y_true, y_pred, y_prob, hyperparams=None):
    cm = confusion_matrix(y_true, y_pred)
    fig, ax = plt.subplots()
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No", "Yes"], yticklabels=["No", "Yes"], ax=ax)
    ax.set_title("Confusion Matrix")
    if hyperparams:
        param_text = "\n".join([f"{k}: {v}" for k, v in hyperparams.items()])
        fig.text(0.99, 0.5, f"Hyperparameters:\n{param_text}", fontsize=8, va='center', ha='left', transform=fig.transFigure,
                 bbox=dict(boxstyle="round", facecolor="wheat", alpha=0.5))
    cm_path = f"{model_name}_cm.png"
    fig.savefig(cm_path, bbox_inches="tight")
    mlflow.log_artifact(cm_path)
    plt.close()

    fpr, tpr, _ = roc_curve(y_true, y_prob)
    fig, ax = plt.subplots()
    ax.plot(fpr, tpr, label="ROC Curve")
    ax.plot([0, 1], [0, 1], linestyle="--")
    ax.set_title("ROC Curve")
    roc_path = f"{model_name}_roc.png"
    fig.savefig(roc_path)
    mlflow.log_artifact(roc_path)
    plt.close()

# ============================
# ✅ TUNING FUNCTION FOR CLASSICAL MODELS
# ============================
def tune_classical_model(model, param_grid, X_train, y_train):
    grid = GridSearchCV(model, param_grid, scoring='roc_auc', cv=3, n_jobs=-1)
    grid.fit(X_train, y_train)
    return grid.best_estimator_, grid.best_params_

# ============================
# ✅ FINAL STEP: DEDUPLICATED PDF REPORT
# ============================
def generate_model_comparison_pdf():
    experiment_id = mlflow.get_experiment_by_name("Churn Models Comparison").experiment_id
    runs = mlflow.search_runs(experiment_ids=[experiment_id])

    # Deduplicate by keeping the best run per model (highest roc_auc)
    runs = runs.sort_values(by="metrics.roc_auc", ascending=False)
    unique_runs = runs.drop_duplicates(subset=["tags.mlflow.runName"], keep="first")

    summary = unique_runs[[
        "tags.mlflow.runName", "metrics.accuracy", "metrics.precision",
        "metrics.recall", "metrics.f1_score", "metrics.roc_auc"
    ]].copy()
    summary.columns = ["model_name", "accuracy", "precision", "recall", "f1_score", "roc_auc"]

    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    pdf.cell(200, 10, txt="Model Comparison Summary", ln=True, align="C")
    pdf.ln(10)

    # Table Header
    col_widths = [50, 25, 25, 25, 25, 25]
    headers = summary.columns.tolist()
    for i, header in enumerate(headers):
        pdf.cell(col_widths[i], 10, header, border=1)
    pdf.ln()

    # Table Rows
    for _, row in summary.iterrows():
        for i, value in enumerate(row):
            val_str = str(round(value, 4)) if isinstance(value, float) else str(value)[:20]
            pdf.cell(col_widths[i], 10, val_str, border=1)
        pdf.ln()

    # Plot Pages
    for _, row in summary.iterrows():
        model_name = row["model_name"]
        for plot_type in ["cm", "roc"]:
            img_path = f"{model_name}_{plot_type}.png"
            if os.path.exists(img_path):
                pdf.add_page()
                pdf.set_font("Arial", size=14)
                pdf.cell(200, 10, txt=f"{model_name} - {plot_type.upper()} Plot", ln=True, align="C")
                pdf.image(img_path, w=180)

    # Output PDF
    pdf_path = "/content/final_model_report.pdf"
    pdf.output(pdf_path)
    files.download(pdf_path)
```
